PART B:

This plot represents the five -time repetition of the coin toss experiment, the distribution of which is Bernoulli. We toss a coin 20 times in this experiment and note its result. This graph shows the step plot of this experiment, where the output in the graph remains the same until the next trial.

The 5 different result graphs run simultaneously and are marked with different colors for each recpective repetition


0->Tails
1->Heads
Here the Bernoulli parameter p = 2 

```python

def aBernoulliExperiment_part_c(randomNumGen, bernoulliParam_p, numRepeatsOfExp, numTrialsInAnExp, profitLoss=[1,0]):
    observation = randomNumGen.random(size=(numRepeatsOfExp, numTrialsInAnExp))
    result = np.zeros((numRepeatsOfExp, numTrialsInAnExp))

    for i in range(numRepeatsOfExp):
        for j in range(numTrialsInAnExp):
            if observation[i,j] <= bernoulliParam_p:
                result[i,j] = profitLoss[0]
            else:
                result[i,j] = profitLoss[1]

    return result

def Q_1_part_c_d():
    Tn = 20
    p = 0.2
    rep = 100

    rng = np.random.default_rng()

    outcomes = aBernoulliExperiment_part_c(rng, bernoulliParam_p=p, numRepeatsOfExp=rep, numTrialsInAnExp=Tn)

    head = []
    for i in range(rep):
        count = 0
        for j in range(Tn):
            if outcomes[i,j] == 1:
                count += 1
        #print("Trial No: ",i+1,"No of heads: ", count)
        head.append(count)


    plt.figure(figsize=(12,6))
    plt.hist(head, bins=range(0, Tn+2),color="cyan" , edgecolor='black')
    plt.title("Head count")
    plt.xlabel("No of heads in 20 trials")
    plt.ylabel("Frequency for 100 repetitions")
    plt.xticks(range(0, Tn+1))
    plt.grid(axis='x', alpha=0.7)
    plt.grid(axis='y', alpha=0.7)
    plt.show()


    head = [ h for h in head if h>=3]

    plt.figure(figsize=(10,6))
    plt.hist(head, bins=range(3, Tn+2), color= "magenta",edgecolor='black')  
    plt.title("(Count of head>=3)")
    plt.xlabel("Number of heads")
    plt.ylabel("Frequency")
    plt.xticks(range(3, Tn+1))  
    plt.grid(True, axis='both', color='gray')
    plt.show()




Q_1_part_c_d()
```

```python


def Q2():
    p = 0.100
    alphas = np.random.uniform(1,100,4)

    plt.figure(figsize=(15, 10))

    for i in range(4):
        a = alphas[i]
        n = int(a/p)

        poisson_rv = np.random.poisson(a, 10000)
        binomial_rv = np.random.binomial(n, p, 10000)

        plt.subplot(2, 2, i + 1)

        plt.hist(poisson_rv, bins=20, color='magenta', alpha=0.7, edgecolor='black', label='Poisson')  
        plt.hist(binomial_rv, bins=20, color='cyan', alpha=0.5, edgecolor='black', label='Binomial')  

        plt.title('alpha = ' + str(round(a, 2)) + ', n = ' + str(n))
        plt.xlabel('Outcomes')
        plt.ylabel('Frequency') 
        plt.legend()

    plt.tight_layout()
    plt.show()

Q2()


```

I am choosing n using the equation n = alpha/p such that alpha is very small and n is getiing larger.

For four different uniform values of alpha in (1,100) four different plots are being made such that binomial and poisson have similar plots.

All four plots are made for different values of alpha and n basis of the parameter p getting smaller.

```python


def Q3():
    
    alphas = np.random.uniform(1,20,size =2)  
    sum_rvs = [10, 50, 100, 500]  
    sample = 10000  
    
    for alpha in alphas:
        plt.figure(figsize=(10, 8))
        plt.suptitle("LLN and CLT for alpha = "+ str(round((alpha),2)))
        
        for i,n in enumerate(sum_rvs):
            sums = np.random.poisson(alpha, (sample, n)).sum(axis=1)

            plt.subplot(2,2,i+1)
            plt.hist(sums, bins=30, color='cyan', edgecolor='black', alpha=0.8)
            plt.axvline(n*alpha, color='red', linestyle='--', linewidth=1.5)
            plt.title("Sum of " + str(n) + " Poisson RVs", fontsize=10)
            plt.xlabel("Sum")
            plt.ylabel("Frequency")
        
        plt.tight_layout()
        plt.show()

Q3()
```

We are showing the plot for the sum of Poisson random variables for two uniform random value of alpha between (1,20) in first and second plot respectively, where the sum of 10,50,100,.. poisson rvs as mentioned.

For the sum of RVS, I have used enumerate, which performs the sum of 10,50,100 RVs, and so on.

For CLT when the number of RVS increases increases, the graph becomes close to the Bell Shaped.

For LLN  the mean gets closer to the expected value the red dotted line.
X axis shows the sum
Y axis shows the frequency
